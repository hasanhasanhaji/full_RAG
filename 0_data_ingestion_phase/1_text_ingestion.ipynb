#### TextLoader --- Read just one document

```python
### Approach 1 to read file -- TextLoader - read a single file
from langchain_community.document_loaders import TextLoader

loader = TextLoader("data/text_files/python_intro.txt", encoding="utf-8")

doc = loader.load()

print(f" Number of documents loaded: {len(doc)}")
print(f" Document content preview: {doc[0].page_content[:200]}...")
print(f" Document metadata: {doc[0].metadata}")
print(f"type of Document: {type(doc)}")
```

#### DirectoryLoader == Raed a whole directory

```python
from langchain_community.document_loaders import DirectoryLoader
```

```python
## load a directory
dir_loader = DirectoryLoader("data/text_files", 
                             glob="*.txt", # optional filter
                             loader_cls = TextLoader,  ## specify the loader class
                             loader_kwargs = {"encoding":"utf-8"}, ## options for the loader class
                             show_progress = True ## show progress bar
                             )

documents = dir_loader.load()

print(f" Total documents loaded: {len(documents)}")
for i,document in enumerate(documents):
    print(f"\n Document {i+1}:")
    print(f" Content preview: {document.page_content[:200]}...")
    print(f" Metadata: {document.metadata}")
    print(f"len of document: {len(document.page_content)} characters")
    print(f"-----------------------------")
    
```

```python
#### print pos and pros of directory approach
print(" DirectoryLoader Pros and Cons:")
print(" Pros:")
print(" - Can load multiple documents from a directory in one go.")
print(" - Flexible with file types and loader classes.")
print(" - Easy to manage and organize document ingestion.")
print(" - progress bar for large directories.")
print(" Cons:")
print(" - May require more memory if loading many large documents.")
print(" - Slightly more complex setup compared to single file loaders.")
print(" - all files must be same type in the directory.")
```

# Text splitting strategies

```python
print(documents)
```

```python
### Method1 --- Character text splitter
text = documents[0].page_content
text
```

```python
print("Method1- CHARACTER TEXT SPLITTER")
char_splitter = CharacterTextSplitter(
    separator= "\n",  ## split on newlines
    chunk_size = 200 , ## max chunk size
    chunk_overlap = 20, ## overlap between chunks
    length_function = len
)

char_chunks = char_splitter.split_text(text)

print(f"Created {len(char_chunks)} chunks")
print(f"First chunk: {char_chunks[0][0:100]}")
```

```python
print(char_chunks)
```

```python
print(char_chunks[2])
print("-------------------")
print(char_chunks[3])
print("-------------------")
```

```python
# we do not have overlap in this method because of new line text splitter that is a clean text splitter
```

```python
## Method 2 --Recursive text splitter ---- the best choice for chunking
recursive_splitter = RecursiveCharacterTextSplitter(
    separators= ["\n\n", "\n", " ", ""],
    chunk_size = 200 , ## max chunk size
    chunk_overlap = 20, ## overlap between chunks
    length_function = len
    
)

recursive_chunks = recursive_splitter.split_text(text)

print(f"Created {len(recursive_chunks)} chunks")
print(f"First chunk: {recursive_chunks[0][0:100]}")
```

```python
print(char_chunks[2])
print("-------------------")
print(char_chunks[3])
print("-------------------")
```

```python
## Because we have \n or \n\n in our text this split does not support overlap.
## if our text does not have \n and we split only on " " (space) so we have overlap.
```

```python
## Method 3 - token text splitter
token_splitter = TokenTextSplitter(
    chunk_size = 50, 
    chunk_overlap = 10)
token_chunks = token_splitter.split_text(text)
print(f"Created {len(token_chunks)} chunks")
print(f"First chunk: {token_chunks[0][0:100]    }")
```
